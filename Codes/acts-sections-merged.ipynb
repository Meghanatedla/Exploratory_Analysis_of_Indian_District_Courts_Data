```python
#Solve DtypeWarning: Columns (X,X) have mixed types. Specify dtype option on import or set low_memory=False in Pandas
acts_sections = pd.read_csv("/kaggle/input/court-data/acts_sections/acts_sections.csv", low_memory=False)
acts_sections.head(10)
```

```python
act_keys = pd.read_csv("/kaggle/input/court-data/keys/keys/act_key.csv")
act_keys.head(10)
```

```python
act_keys.drop(columns=['count'], inplace=True)
```

```python
acts_sections = pd.merge(acts_sections, act_keys, how='left', on='act')
acts_sections.head(10)
```

```python
section_keys = pd.read_csv("/kaggle/input/court-data/keys/keys/section_key.csv")
section_keys.head(10)
```

```python
section_keys.drop(columns=['count'], inplace=True)
```

```python
acts_sections = pd.merge(acts_sections, section_keys, how='left', on='section')
acts_sections.head(10)
```

```python
acts_sections.to_csv("/kaggle/working/acts_sections_merged.csv", index=False)
```
